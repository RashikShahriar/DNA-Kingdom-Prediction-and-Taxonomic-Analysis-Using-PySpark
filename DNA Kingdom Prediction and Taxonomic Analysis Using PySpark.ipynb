```python
#Handling Missing value for string data using maximum frequency category
column_subset = [col_ for col_ in dataframe.columns if dataframe.select(col_).dtypes[0][1] =="string"]
for col_ in column_subset:
    temp_col = dataframe.groupBy(col_).count()
    temp_col = temp_col.dropna(subset=col_)
    frequent_category = temp_col.orderBy(
                        temp_col['count'].desc()).collect()[0][0]
    dataframe=dataframe.na.fill(frequent_category, col_)
```

# Z-score calculation (Normal Method)

```python
from pyspark.sql.functions import col, mean, stddev, expr

# Assuming 'indexed_dataframe' is your DataFrame
df = indexed_dataframe

print(df.schema.fields)

# Get a list of numeric columns
numeric_cols = [field.name for field in df.schema.fields if str(field.dataType) in ['IntegerType()', 'LongType()', 'FloatType()', 'DoubleType()']]

print(numeric_cols)

# Calculate mean and stddev for each numeric column
stats = df.select(
    *[mean(col).alias(f'{col}_mean') for col in numeric_cols],
    *[stddev(col).alias(f'{col}_stddev') for col in numeric_cols]
).collect()[0]

# Compute z-scores for each column
for col_name in numeric_cols:
    mean_col = stats[f'{col_name}_mean']
    stddev_col = stats[f'{col_name}_stddev']
    df = df.withColumn(f'{col_name}_zscore', (col(col_name) - mean_col) / stddev_col)

# Show the resulting DataFrame with z-scores
df.show()
```

# Z-score calculation using rdd mapping

```python
from pyspark.sql.functions import mean, stddev

# Assuming your DataFrame is named 'dataframe'
columns = indexed_dataframe.columns

# Compute mean and standard deviation for each column
stats = indexed_dataframe.select([mean(col).alias(col + '_mean') for col in columns] +
                         [stddev(col).alias(col + '_stddev') for col in columns]).collect()[0]

# Create dictionaries for means and standard deviations
means = {col: stats[col + '_mean'] for col in columns}
stddevs = {col: stats[col + '_stddev'] for col in columns}

# Convert DataFrame to RDD
rdd = indexed_dataframe.rdd

# Define a function to compute z-scores
def compute_z_scores(row):
    z_scores = {}
    for col in columns:
        mean_val = means[col]
        stddev_val = stddevs[col]
        z_scores[col] = (row[col] - mean_val) / stddev_val if stddev_val != 0 else None
    return tuple(z_scores[col] for col in columns)

# Apply the function to each row
z_scores_rdd = rdd.map(compute_z_scores)

# Convert back to DataFrame if needed
z_scores_df = z_scores_rdd.toDF(columns)

# Show the resulting DataFrame
z_scores_df.show()
```

# IQR calculation using RDD Mapping




```python
# Get column names
columns = indexed_dataframe.columns

# Convert DataFrame to RDD
rdd = indexed_dataframe.rdd

# Calculate IQR for each column
for column in columns:
    # Extract column values
    column_rdd = rdd.map(lambda row: row[column])

    # Sort the RDD
    sorted_rdd = column_rdd.sortBy(lambda x: x)

    # Calculate the quartiles
    count = sorted_rdd.count()
    Q1_index = int(0.25 * count)
    Q3_index = int(0.75 * count)

    Q1 = sorted_rdd.zipWithIndex().filter(lambda x: x[1] == Q1_index).map(lambda x: x[0]).collect()[0]

    Q3 = sorted_rdd.zipWithIndex().filter(lambda x: x[1] == Q3_index).map(lambda x: x[0]).collect()[0]

    # Calculate IQR
    IQR = Q3 - Q1

    print(f"Interquartile Range (IQR) for '{column}':", IQR)
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml import Pipeline

# Assemble features into a vector
feature_columns = [col for col in indexed_dataframe.columns if col != 'kingdom_indexed']
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
assembled_df = assembler.transform(indexed_dataframe)

# Split the data into training and testing sets
(training_data, testing_data) = assembled_df.randomSplit([0.7, 0.3], seed=42)

# Train the models
lr = LogisticRegression(labelCol="kingdom_index", featuresCol="features")

# Fit models
lr_model = lr.fit(training_data)

# Evaluate the models
evaluator = MulticlassClassificationEvaluator(labelCol="kingdom_index", predictionCol="prediction", metricName="accuracy")

# Define a function for evaluation
def evaluate_model(model):
    predictions = model.transform(testing_data)
    accuracy = evaluator.evaluate(predictions)
    return accuracy

# Evaluate each model
lr_accuracy = evaluate_model(lr_model)

# Print accuracies
print("Logistic Regression Accuracy:", lr_accuracy)
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import RandomForestClassifier, DecisionTreeClassifier, LinearSVC, OneVsRest
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Step 1: Copy the dataframe
codons_df = indexed_dataframe

# Step 2: Remove specified columns
columns_to_remove = ["class_index", "speciesname_index", "superkingdom_index", "clade_index",
                     "family_index", "type_index", "no_rank_index", "phylum_index",
                     "order_index", "genus_index", "species_index", "species_subgroup_index",
                     "forma_specialis_index"]
codons_df = codons_df.drop(*columns_to_remove)

# Step 3: Prepare the data
feature_columns = [col for col in codons_df.columns if col != 'kingdom_index']
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
codons_df = assembler.transform(codons_df)

# Split the data
train_df, test_df = codons_df.randomSplit([0.7, 0.3], seed=42)

# Initialize evaluator
evaluator = MulticlassClassificationEvaluator(labelCol="kingdom_index", predictionCol="prediction", metricName="accuracy")

# Random Forest
rf = RandomForestClassifier(labelCol="kingdom_index", featuresCol="features")
rf_model = rf.fit(train_df)
rf_predictions = rf_model.transform(test_df)
rf_accuracy = evaluator.evaluate(rf_predictions)
print(f"Random Forest Accuracy: {rf_accuracy}")

# Decision Tree
dt = DecisionTreeClassifier(labelCol="kingdom_index", featuresCol="features")
dt_model = dt.fit(train_df)
dt_predictions = dt_model.transform(test_df)
dt_accuracy = evaluator.evaluate(dt_predictions)
print(f"Decision Tree Accuracy: {dt_accuracy}")

# SVM with OneVsRest
svm = LinearSVC(labelCol="kingdom_index", featuresCol="features")
ovr = OneVsRest(classifier=svm, labelCol="kingdom_index", featuresCol="features")
ovr_model = ovr.fit(train_df)
ovr_predictions = ovr_model.transform(test_df)
ovr_accuracy = evaluator.evaluate(ovr_predictions)
print(f"SVM Accuracy (OneVsRest): {ovr_accuracy}")
```
